```python
#libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
 
# Values of each group
Jan = [15,13,14,22,0]
Feb = [20,15,22,17,0]
March = [20,11,22,18,0]
April = [12,17,14,17,0]
May = [7,13,17,10,0]
June= [10,10,17,10,0]
July=[7,18,17,19,0]
August=[18,21,17,14,0]
September=[10,8,14,17,0]
October=[20,22,17,11,0]
November=[16,12,13,11,0]
December=[19,15,11,12,0]

# The position of the bars on the x-axis
r = [0,1,2,3,6]
 
# Names of group and bar width
names = ['2011','2012','2013','2014', '']
barWidth = 0.8
 
barsMarch = np.add(Jan,Feb).tolist()
barsApril = np.add(barsMarch, March).tolist()

barsMay = np.add(barsApril, April).tolist()
barsJune = np.add(barsMay, May).tolist()
barsJuly = np.add(barsJune, June).tolist()
barsAugust = np.add(barsJuly, July).tolist()
barsSeptember = np.add(barsAugust, August).tolist()
barsOctober = np.add(barsSeptember, September).tolist()
barsNovember = np.add(barsOctober, October).tolist()
barsDecember = np.add(barsNovember, November).tolist()

plt.bar(r, Jan, label='Jan',color='skyblue', edgecolor='white', width=barWidth)
plt.bar(r, Feb, bottom=Jan, label='Feb', color='orange', edgecolor='white', width=barWidth)
plt.bar(r, March, bottom=barsMarch, label='March', color='gray', edgecolor='white', width=barWidth)
plt.bar(r, April, bottom=barsApril, label='April', color='black', edgecolor='white', width=barWidth)
plt.bar(r, May, bottom=barsMay, label='May', color='lightgreen', edgecolor='white', width=barWidth)
plt.bar(r, June, bottom=barsJune, label='June', color='green', edgecolor='white', width=barWidth)
plt.bar(r, July, bottom=barsJuly, label='July', color='Darkgreen', edgecolor='white', width=barWidth)
plt.bar(r, August, bottom=barsAugust, label='August', color='Darkgray', edgecolor='white', width=barWidth)
plt.bar(r, September, bottom=barsSeptember, label='September', color='salmon', edgecolor='white', width=barWidth)
plt.bar(r, October, bottom=barsOctober, label='October', color='cyan', edgecolor='white', width=barWidth)
plt.bar(r, November, bottom=barsNovember, label='November', color='gold', edgecolor='white', width=barWidth)
plt.bar(r, December, bottom=barsDecember, label='December', color='Silver', edgecolor='white', width=barWidth)
 
# Custom X axis
plt.xticks(r, names, fontweight='bold')
plt.xlabel("Total Robberies - Karachi")
plt.ylabel("Number of Robberies")
plt.title("Stack Map of Karachi Robberies 2011-2014")
plt.legend(loc= "upper right", fancybox=True, framealpha=1, shadow=True, borderpad=1)

# Show graphic
plt.show()
```
