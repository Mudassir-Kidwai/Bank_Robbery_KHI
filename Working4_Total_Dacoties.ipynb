```python
import matplotlib.pyplot as plt

xbars=[2011,2012,2013,2014]
bars=["2011","2012","2013","2014"]
height=[174, 169, 195, 178]
width=[0.5,0.5,0.5,0.5]
label=[]
plt.bar(xbars,height,width=width, color =['red','blue','green','yellow'],edgecolor='black')
plt.xlabel("YEAR")
plt.ylabel("Total Dacoity/Robbery")
plt.xticks(xbars, bars)
plt.title("Total Robberies by Year 2011-2014 in KHI")
#plt.legend()

# Create labels
label = ['174','169','195','178']
# Text on the top of each barplot
for i in range(len(height)):
    plt.text(x = xbars[i]-0.1 , y = height[i]+0.7, s = label[i], size = 8)
 

plt.show()
```
