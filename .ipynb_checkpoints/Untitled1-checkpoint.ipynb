```python
import matplotlib.pyplot as plt

xbars=[2011,2012,2013,2014]
bars=["2011","2012","2013","2014"]
height=[106, 86, 100, 105]
width=[0.5,0.5,0.5,0.5]
label=[]
plt.bar(xbars,height,width=width, color =['red','blue','green','yellow'],edgecolor='black')
plt.xlabel("YEAR")
plt.ylabel("Murders during Dacoity/Robbery")
plt.xticks(xbars, bars)
plt.title("Murders during Dacoity/Robbery by Year 2011-2014 in KHI")
#plt.legend()

# Create labels
label = ['106','86','100','105']
# Text on the top of each barplot
for i in range(len(height)):
    plt.text(x = xbars[i]-0.1 , y = height[i]+0.3, s = label[i], size = 8)
 

plt.show()
```
